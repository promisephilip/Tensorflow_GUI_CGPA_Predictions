```python
#Importing Library
import tkinter as tk
import pickle
import os
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import joblib
import os.path

#load model
model = pickle.load(open(r"C:\Users\hp 4430s\Desktop\Tensorflow_prediction\Studentmodel.pkl", 'rb'))

#Craeting a new tkinter window
main = tk.Tk() 

# assigning a title 
main.title("STUDENT DATA ANALYSIS - SDA") 

# specifying geomtery for window size 
main.geometry("650x300")

#Defining entry variables
num1 = DoubleVar()
num2 = DoubleVar()
num3 = DoubleVar() 

e_1 = StringVar()
e_2 = StringVar()
e_3 = StringVar()
e_4 = StringVar()

#float
cgpa1 = float(num1.get())
cgpa2 = float(num2.get())
cgpa3 = float(num3.get())

#functions
def about():
    messagebox.showinfo('STUDENT DATA ANALYSIS - SDA','This program predicts the CGPA of the fourth year of any university student with a maximum cgpa of 5.0.\n\n This program also saves the inputed details to a txt file')
    
def output():
    global cgpa4
    #float
    cgpa1 = float(num1.get())
    cgpa2 = float(num2.get())
    cgpa3 = float(num3.get())
    cgpa4 = model.predict([[cgpa1, cgpa2, cgpa3]])
    tk.Label(main, text=str(cgpa4)).grid(row=14, column=1)

def clear():
    global c_4
    global save_name
    os.system('cls')
    num1.set(' ')
    num2.set(' ')
    num3.set(' ')
    e_1.set(' ')
    e_2.set(' ')
    e_3.set(' ')
    e_4.set(' ')
    c_4 = tk.Label(main, text=' '*30).grid(row=14, column=1)
    save_name = tk.Label(main, text= ' '*50).grid(row=15, column = 1)
    
def save():
    first_name = e1.get()
    last_name= e2.get()
    reg_num = e3.get()
    exam_number = e4.get()
    cg1 = str(float(num1.get()))
    cg2 = str(float(num2.get()))
    cg3 = str(float(num3.get()))
    cg4 = str(cgpa4)
    
    save_path = r"C:\Users\hp 4430s\Desktop\Tensorflow_prediction"
    
    saving = os.path.join(save_path, first_name+'_'+exam_number+".txt")
    
    file_object = open(saving, "w")
    file_object.write('First Name: '+first_name)
    file_object.write('\n')
    file_object.write('Last Name: '+last_name)
    file_object.write('\n')
    file_object.write('Registration Number: '+reg_num)
    file_object.write('\n')
    file_object.write('Examination Number: '+exam_number)
    file_object.write('\n')
    file_object.write('CGPA-1: '+cg1)
    file_object.write('\n')
    file_object.write('CGPA-2: '+cg2)
    file_object.write('\n')
    file_object.write('CGPA-3: '+cg3)
    file_object.write('\n')
    file_object.write('CGPA-4: '+cg4)
        
    savedas = 'File has been saved as '+ first_name +'_'+ exam_number
    save_name = tk.Label(main, text= savedas).grid(row=15, column = 1)
    file_object.close()
    
#Menu
menu = Menu(main)
main.config(menu=menu)

filemenu = Menu(menu,background='#ff8000',
                foreground='black', 
                activebackground='white', 
                activeforeground='black')
menu.add_cascade(label='File',
                 menu=filemenu,
                 background='#ffcc99',
                underline=0)
filemenu.add_command(label='Save', command= lambda : save())
filemenu.add_separator()
filemenu.add_command(label='Exit', command=main.destroy)

helpmenu = Menu(menu)
menu.add_cascade(label='Help', menu=helpmenu)
helpmenu.add_command(label='About', command=about)

# label to enter first name 
tk.Label(main, text="First Name: ").grid(row=0, column=0) 
e1 = tk.Entry(main, textvariable = e_1)
e1.grid(row=0, column=1) 

# label to enter last name 
tk.Label(main, text="Last Name: ").grid(row=1, column=0) 
e2 = tk.Entry(main, textvariable = e_2)
e2.grid(row=1, column=1) 

# label for registration number 
tk.Label(main, text="Matric No: ",).grid(row=0, column=4)
e3 = tk.Entry(main, textvariable = e_3)
e3.grid(row=0, column=6)

# label for examination Number 
tk.Label(main, text="Examination No: ").grid(row=1, column=4)
e4 = tk.Entry(main, textvariable = e_4)
e4.grid(row=1, column=6)

label = tk.Label(main, text="Enter CGPA for Year 1, Year 2 and Year 3 below.", font=("Arial Bold", 10)).grid(row=4, column=1)

#year 1 CGPA entry
tk.Label(main, text="Year 1").grid(row=5, column=1) 
tk.Label(main, text="CGPA 1:").grid(row=6, column=0)
cgpa1 = Entry(main, textvariable = num1)
cgpa1.grid(row=6, column=1)
 
#year 2 CGPA entry
tk.Label(main, text="Year 2").grid(row=7, column=1) 
tk.Label(main, text="CGPA 2").grid(row=8, column=0) 
cgpa2 = Entry(main, textvariable = num2)
cgpa2.grid(row=8, column=1)

#year 3 CGPA entry
tk.Label(main, text="Year 3").grid(row=9, column=1) 
tk.Label(main, text="CGPA 3:").grid(row=10, column=0)
cgpa3 = Entry(main, textvariable = num3)
cgpa3.grid(row=10, column=1)

#year 4 CGPA entry
cgpa4_input = tk.Label(main, text="Year 4").grid(row=13, column=1) 
cgpa_4 = tk.Label(main, text="CGPA 4:").grid(row=14, column=0)

#button to predict and display year 4 CGPA
button1 = tk.Button(main, text="Predict", 
                    bg="Lightblue",
                    fg="black",
                    command=output)
button1.grid(row=12, column=1)

button2 = tk.Button(main, text="Clear", 
                    command=clear)
button2.grid(row=12, column=2)
    
main.mainloop()
```
